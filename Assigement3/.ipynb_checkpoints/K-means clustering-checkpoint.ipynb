```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Read the data from the CSV file
data = pd.read_csv("Mall_Customers.csv")
```

```python
# Select the columns that will be used for clustering
X = data[["Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
# Specify the number of clusters (K)
kmeans = KMeans(n_clusters=4)
```

```python
# Fit the model to the data
kmeans.fit(X)
```

```python
# Get the cluster labels for each data point
data["Cluster"] = kmeans.labels_
```

```python
# Print the first few rows of the data with the cluster labels
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the data with different colors for each cluster
plt.scatter(data[data["Cluster"] == 0]["Age"], data[data["Cluster"] == 0]["Annual Income (k$)"], color = "red", label = "Cluster 1")
plt.scatter(data[data["Cluster"] == 1]["Age"], data[data["Cluster"] == 1]["Annual Income (k$)"], color = "blue", label = "Cluster 2")
plt.scatter(data[data["Cluster"] == 2]["Age"], data[data["Cluster"] == 2]["Annual Income (k$)"], color = "green", label = "Cluster 3")
plt.scatter(data[data["Cluster"] == 3]["Age"], data[data["Cluster"] == 3]["Annual Income (k$)"], color = "cyan", label = "Cluster 4")

# Add the cluster centroids to the plot as well
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = "black", marker = "X", label = "Centroids")

# Add labels and title to the plot
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.title("Clusters of Mall Customers")
plt.legend()

# Show the plot
plt.show()
```
